```python
import json

#f = open('yelp_academic_dataset_business.json',)
data = [json.loads(line) for line in open('yelp_academic_dataset_business.json', 'r')]
```

```python
len(data)
```

```python
states = {}

for i in range(len(data)):
    state = data[i]['state'].lower()
    if state not in states.keys():
        states[state] = 1
    else:
        states[state] +=1
        
states
```

```python
cities = {}

for i in range(len(data)):
    city = data[i]['city'].lower()
    if city not in cities.keys():
        cities[city] = 1
    else:
        cities[city] +=1
cities
```

```python
bc = {}

for i in range(len(data)):
    state = data[i]['state'].lower()
    city = data[i]['city'].lower()
    
    if state == 'bc':
        if city not in bc.keys():
            bc[city] = 1
        else:
            bc[city] +=1
bc
```

```python
s = 0
for key in bc:
    if 'vancouver' in key:
        s = s + bc[key]
s
```

```python
business_data = []

for i in range(len(data)):
    city = data[i]['city'].lower()
    state = data[i]['state'].lower()
    
    if state == 'bc' and ('vancouver' in city):
        business_data.append(data[i])

len(business_data)
```

```python
business_id_lst = []
for i in range(len(business_data)):
    if business_data[i]['business_id'] not in business_id_lst:
        business_id_lst.append(business_data[i]['business_id'])
len(business_id_lst)
```
